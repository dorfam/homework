```python
c=pres.totincr.value_counts().sort_index()
thinkstats2.Hist(c)
```

```python
neverMarried=pres[pres.fmarno==0]
married=pres[pres.fmarno >=1]
ht= thinkstats2.Hist(neverMarried.totincr)
ht2=thinkstats2.Hist(married.totincr)
import thinkplot
width=0.25
thinkplot.Hist(ht,align="right", width=width)
thinkplot.Hist(ht2,align="left", width=width)
thinkplot.show(xlabel="weeks", ylabel="count")
```

```python
print ("the minummum value of the never married's income is %.2f" %(neverMarried.totincr.min()))
print ("the minummum value of the married's income is %.2f" %(married.totincr.min()))
print ("the maximum value of the never married's income is %.2f" %(neverMarried.totincr.max()))
print ("the maximum value of the  married's income is %.2f" %(married.totincr.max()))
print ("the standard deviation of the never married's income is %.2f" %(neverMarried.totincr.std()) )
print ("the standard deviation of the married's income is %.2f" %(married.totincr.std()))
print ("the variance of the never married's income is %.2f" %(neverMarried.totincr.var()) )
print ("the variance of the married's income is %.2f" %(married.totincr.var()) )
print ("the mean of the never married's income is %.2f" %(neverMarried.totincr.mean()) )
print ("the mean of the  married's income is %.2f" %(married.totincr.mean()) )
```
